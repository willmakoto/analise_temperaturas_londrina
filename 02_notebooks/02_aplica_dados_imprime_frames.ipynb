```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import os
from pathlib import Path
```

```python
# Plota a gráfico polar
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(111, projection='polar')

# Apaga as linhas e marcações
ax1.grid(False)
ax1.axes.get_yaxis().set_ticklabels([])
ax1.axes.get_xaxis().set_ticklabels([])

# Muda as cores do gráfico
fig.set_facecolor("black")
ax1.set_facecolor("black")

# Coloca título
ax1.set_title("Média mensal da temperatura máxima \n na cidade de Londrina/PR (1961-2019)", color='white', fontdict={'fontsize': 30}, y= 1.1, loc = "center")

# Seta 0 graus na posição de cima e seta clockwise
ax1.set_theta_zero_location("N")
ax1.set_theta_direction(-1)

# Coloca os meses nas devidas posições
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dec"]
dist = (1.1,1.1,1.12,1.13,1.14,1.15,1.13,1.14,1.13,1.12,1.1,1.1)
i = 0
for i in range(0,12,1):
    ax1.text(i*2*np.pi/12 ,dist[i] , meses[i], color='white', size=20, ha='center')


# Circulos e temperaturas de referência
full_circle_thetas = np.linspace(0, 2*np.pi, 1000)
temp = [35, 30, 25, 20, 15]
raio = [1.0, 0.8, 0.6, 0.4, 0.2]

for t, r in zip(temp, raio):
    ax1.plot(full_circle_thetas, [r]*len(full_circle_thetas), c='white')
    ax1.text(2.5*2*np.pi/12, r,
        f"{t}°C" if t == 35 else f"{t}", color="white", fontsize=18)

ano = 1961
frames = []
frame_count = 0
df1 = pd.read_csv("../01_dados/processados/dados_convertidos.csv")

for ano in df1["Ano"].unique():
    if ano == 2020:
        break
    # Tira o ano anterior do centro
    for txt in ax1.texts:
        if(txt.get_position()==(6*2*np.pi/12,0.03)):
            txt.set_visible(False)
            
    # Coloca o ano no centro
    ax1.text(6*2*np.pi/12, 0.03, ano, color='white', size=20, ha='center')

    df1_parcial = df1[df1['Ano'] == ano]
    
    for mes in df1_parcial["Mes_deslocado"]:
        x1 = np.random.uniform(mes * 2 * np.pi / 12 - 2 * np.pi / 48, 
                               mes * 2 * np.pi / 12 + 2 * np.pi / 48)
        y1 = df1_parcial.loc[df1_parcial["Mes_deslocado"] == mes, "Tmax_conv"].values[0]
        if 0.8 > y1 > 0.6:
            ax1.scatter(x1, y1, s=100, color="yellow", alpha=0.5)
        elif y1 <= 0.6:
            ax1.scatter(x1, y1, s=100, color="blue", alpha=0.5)
        elif y1 >= 0.8:
            ax1.scatter(x1, y1, s=100, color="red", alpha=0.5)

        # pasta onde o notebook está
        BASE_DIR = Path.cwd().parent

        # cria pasta para os frames
        frames_dir = BASE_DIR  / "03_outputs" / "frames"
        frames_dir.mkdir(parents=True, exist_ok=True)
        frame_path = frames_dir / f"frame_{frame_count}.png"
        
        # salva os frames
        plt.savefig(frame_path, facecolor=fig.get_facecolor(), bbox_inches="tight")

        frames.append(frame_path)
        frame_count += 1
```
