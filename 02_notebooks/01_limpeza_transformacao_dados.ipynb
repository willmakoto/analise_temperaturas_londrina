```python
import pandas as pd
import glob
```

```python

def junta_csv():
    # Coloca todos os arquivos CSV de uma pasta em uma lista
    csv_files = glob.glob('../01_dados/brutos/*.csv')

    # Cria DataFrame vazio
    combined_df = pd.DataFrame()

    # Coloca todos CSV no DataFrame combined_df
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        combined_df = pd.concat([combined_df, df], ignore_index=True)

    # Transforma o DataFrame em um novo arquivo CSV
    combined_df.to_csv("../01_dados/brutos/bruto_tudo.csv")


junta_csv()
```

```python
def calcular_media_tmax(arquivo_entrada, arquivo_saida):
    # Carrega os dados
    df = pd.read_csv(arquivo_entrada)
    
    # Converte a coluna 'system:time_start' para datetime
    df["system:time_start"] = pd.to_datetime(df["system:time_start"])
    
    # Cria uma nova coluna com o ano e mês
    df["Ano_Mes"] = df["system:time_start"].dt.to_period("M")
    
    # Calcula a média de Tmax por mês
    media_mensal = df.groupby("Ano_Mes")["Tmax"].mean()
    
    # Salva o resultado em um novo arquivo CSV
    media_mensal.to_csv(arquivo_saida, header=["Media_Tmax"])


calcular_media_tmax("../01_dados/brutos/bruto_tudo.csv", "../01_dados/processados/media_tmax.csv")
```

```python
def transformando_df(arquivo_entrada, arquivo_saida):
    df1 = pd.read_csv(arquivo_entrada)

    #Cria colunas Ano e Mes
    df1["Ano"] = df1["Ano_Mes"].str.split('-').str[0]
    df1["Mes"] = df1["Ano_Mes"].str.split('-').str[1]

    #Reescala coluna Tmax
    def R(M):
        R = 0.2 + 0.04*(M-15)
        return R
    df1["Tmax_conv"] = df1["Media_Tmax"].map(R)
    
    #Cria coluna Mes_deslocado
    df1["Mes_deslocado"] = df1["Mes"].astype(int) -1

    #Dropa coluna Ano_Mes e Media_Tmax
    df1 = df1.drop(columns = ["Ano_Mes", "Media_Tmax"])

    # Salva DataFrame em arquivo CSV
    df1.to_csv(arquivo_saida)


transformando_df("../01_dados/processados/media_tmax.csv", "../01_dados/processados/dados_convertidos.csv")
    
```
